```python
for f in range(len(all_filenames)):
    try:
        df=pd.read_csv(all_filenames[f])
    except:
        df=pd.read_csv(all_filenames[f],encoding='latin-1')
    df.columns= df.columns.str.lower()
    for i in range(len(df.columns)-1):
        if 'first' in str(df.columns[i]):
            elver=pd.DataFrame()
            elver=df.iloc[:,[i]]
            elver.replace(np.nan,'Colleague',inplace=True)
    for i in range(len(df.columns)-1):
        if 'email' in str(df.columns[i]):
            elver['Email']=df.iloc[:,[i]]
    #elver.set_index(elver.iloc[0].values,inplace=True)
    elver.to_csv(all_filenames[f])
```
